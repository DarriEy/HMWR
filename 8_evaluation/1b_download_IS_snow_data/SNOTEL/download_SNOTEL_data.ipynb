```python
import os
import glob
import sys
import urllib3
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
pd.options.mode.chained_assignment = None
```

## Define function to download data

```python
def getSnotelData(SiteID, StateAbb, SiteName, root_path):
	
	url1 = 'https://wcc.sc.egov.usda.gov/reportGenerator/view_csv/customSingleStationReport/daily/'
	url2 = str(SiteID) + ':' + StateAbb +':SNTL%7Cid=%22%22%7Cname/'
	url2b = 'POR_BEGIN,POR_END/WTEQ::value,PREC::value,PRCP::value'
	url = url1+url2+url2b
	
	print(f'Start retrieving data for {SiteName}, {SiteID}')
	
	http = urllib3.PoolManager()
	response = http.request('GET', url)
	data = response.data.decode('utf-8')

	filename = 'SNOTEL_Site_' + SiteName + '_' + str(SiteID) + '.csv' 
	filename.replace(' ', '_')
	
	with open(root_path + filename, 'w') as x_file:
		
		x_file.write(data)
```

## Read the file that contains information about SNOTEL sites

```python
SNOTEL_info = pd.read_csv('./NRCS_SNOTEL_Locations_wAlaska.csv', encoding='latin-1')
```

## Loop over the SNOTEL locations in the file
